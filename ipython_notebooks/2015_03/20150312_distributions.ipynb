```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
sample1 = np.random.randn(100)*10 - 7.5
sample2 = np.random.randn(100)*10 + 7.5
```

```python
def plot_dists(ax, samp1, samp2, bound=50, width=5, norm=True):
    """Makes the distribution plots"""
    # Gets the histograms
    hist1, bin_ = np.histogram(samp1, np.arange(-bound, bound+1, width))
    hist2, bin_ = np.histogram(samp2, np.arange(-bound, bound+1, width))
    
    bin_ = np.arange(-bound, bound, width)
    
    # Normalizes the histogram
    hist1 = hist1/float(hist1.max())
    hist2 = hist2/float(hist2.max())
    
    # Plots the first distribution as a histogram
    ax.bar(bin_, hist1*0.75, width=width, bottom=0.125, color='#a6cee3', edgecolor='#1f78b4')
    ax.bar(bin_, bin_*0, width=width, bottom=0.125, color='w', edgecolor='w')
    # Plots the second distribution as a histogram
    ax.bar(bin_, hist2*0.75, width=width, bottom=1.125, color='#fb9a99', edgecolor='#e31a1c')
    ax.bar(bin_, bin_*0, width=width, bottom=1.125, color='w', edgecolor='w')

    # Sets the axis limits
    ax.set_xlim([-bound, bound])
    ax.set_ylim([0, 2])
    # Hides the vertical box
    ax.yaxis.set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # Displays the y axis correctly
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_color('k')
    ax.spines['bottom'].set_linewidth(2)
    ax.xaxis.set_ticks_position('bottom')
    ax.xaxis.set_tick_params(direction='out', color='k', width=2)
    ax.set_xticklabels('')
    # Colors the axis
    ax.set_axis_bgcolor('w')
```

```python
fig = plt.figure()
fig.set_facecolor('w')
ax0 = fig.add_axes([0.125, 0.125, 0.7, 0.4])
ax1 = fig.add_axes([0.125, 0.600, 0.2, 0.2])
ax2 = fig.add_axes([0.375, 0.600, 0.2, 0.2])
ax3 = fig.add_axes([0.625, 0.600, 0.2, 0.2])
plot_dists(ax0, sample1, sample2)
plot_dists(ax1, np.random.choice(sample1, 10, replace=False),  np.random.choice(sample2, 10, replace=False))
plot_dists(ax2, np.random.choice(sample1, 10, replace=False),  np.random.choice(sample2, 10, replace=False))
plot_dists(ax3, np.random.choice(sample1, 10, replace=False),  np.random.choice(sample2, 10, replace=False))
fig.savefig('/Users/jwdebelius/Desktop/subsample.png', facecolor='none')
```

```python
def plot_dists2(ax, samp1, samp2, bound=50, width=5, norm=True):
    """Makes the distribution plots"""
    # Gets the histograms
    hist1, bin_ = np.histogram(samp1, np.arange(-bound, bound+1, width))
    hist2, bin_ = np.histogram(samp2, np.arange(-bound, bound+1, width))
    
    bin_ = np.arange(-bound, bound, width)
    
    # Normalizes the histogram
    hist1 = hist1/float(hist1.max())
    hist2 = hist2/float(hist2.max())
    
    # Plots the first distribution as a histogram
    ax.bar(bin_, hist1, width=width, bottom=0, 
           color='#a6cee3', edgecolor='#1f78b4')
    # Plots the second distribution as a histogram
    ax.bar(bin_, hist2, width=width, bottom=0, 
           color='#e31a1c', edgecolor='#fb9a99')
    # Plots the overlap
    ax.bar(bin_, np.vstack((hist1, hist2)).min(0), 
           width=width, bottom=0, 
           color='#cab2d6', edgecolor='#6a3d9a')
    
    # Adjusts the background
    ax.set_axis_bgcolor('w')
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
fig2 = plt.figure()
# fig2.set_facecolor('k')
ax20 = fig2.add_axes([0.125, 0.125, 0.7, 0.4])
ax21 = fig2.add_axes([0.125, 0.600, 0.2, 0.2])
ax22 = fig2.add_axes([0.375, 0.600, 0.2, 0.2])
ax23 = fig2.add_axes([0.625, 0.600, 0.2, 0.2])

plot_dists2(ax20, sample1, sample2)
plot_dists2(ax21, np.random.choice(sample1, 10, replace=False), np.random.choice(sample2, 10, replace=False))
plot_dists2(ax22, np.random.choice(sample1, 10, replace=False), np.random.choice(sample2, 10, replace=False))
plot_dists2(ax23, np.random.choice(sample1, 10, replace=False), np.random.choice(sample2, 10, replace=False))
ax21.set_ylim([0, 1.25])
ax22.set_ylim([0, 1.25])
ax23.set_ylim([0, 1.25])
fig.set_size_inches((8, 4))
```
